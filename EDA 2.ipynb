```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Problem Statement
### A retail company 'ABC Private Limited' wants to understand the customer purchase behaviour (specifically, purchase amount) against various product of differnt categories. They have shared purchase summary of various customers for selected high volume products from ast month. The data set also contains customer demographics (age, gender, martial status, city_type, stay_in_current_city), product details(product_id and product_category) and Total purchase_amount from last month.
### Now they want to build a model to predict the purchase amount of customer against various products which will help them to create personalised offer for customers against different products.

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
df['Gender'] = df['Gender'].map({'F':0, 'M':1})
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
  
df['Age'].unique()
```

```python
df.head()
```

```python
df_City = pd.get_dummies(df['City_Category'], drop_first = True)
df_City.head()
```

```python
df = pd.concat([df,df_City],axis =1)
df.head()
```

```python
df.drop('City_Category',axis = 1, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Product_Category_1'].unique()
```

```python
df['Product_Category_1'].mode()[0]
```

```python
df['Product_Category_1'] = df['Product_Category_1'].fillna(df['Product_Category_1'].mode()[0])
```

```python
df['Product_Category_1'].isnull().sum()
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_1'].isnull().sum()
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
# convert object into integers
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python
df['B'] = df['B'].astype(int)
df['C'] = df['C'].astype(int)
```
